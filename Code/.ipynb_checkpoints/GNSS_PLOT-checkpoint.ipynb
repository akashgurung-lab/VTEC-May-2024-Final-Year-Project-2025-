```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# --- 1. Final Station Data: Using the coordinates you provided ---
station_coords = {
    'ALGO': {'lat': 45.95580, 'lon': -78.07137},  # 281.92863 - 360
    'AMC4': {'lat': 38.80312, 'lon': -104.52459}, # 255.47541 - 360
    'CHUR': {'lat': 58.75908, 'lon': -94.08873},  # 265.91127 - 360
    'IRKJ': {'lat': 52.21903, 'lon': 104.31619},
    'PIE1': {'lat': 34.30151, 'lon': -108.11893}, # 251.88107 - 360
    'SANT': {'lat': -33.15029, 'lon': -70.66855},  # 289.33145 - 360
    'SHAO': {'lat': 31.09964, 'lon': 121.20044},
    'ZECK': {'lat': 43.78841, 'lon': 41.56508},
    'ZIMM': {'lat': 46.87709, 'lon': 7.46527}
}
df_stations = pd.DataFrame.from_dict(station_coords, orient='index')


# --- 2. Create the Enhanced Map Plot ---
print("Generating final map with corrected coordinates...")

projection = ccrs.Robinson(central_longitude=0)
transform = ccrs.PlateCarree()

fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={'projection': projection})
ax.set_global()

# --- 3. Add Scientific Map Features ---
ax.add_feature(cfeature.LAND, facecolor='#f0f0f0', zorder=0)
ax.add_feature(cfeature.OCEAN, facecolor='#d9eaf7')
ax.add_feature(cfeature.COASTLINE, edgecolor='gray', linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='gray')
ax.add_feature(cfeature.LAKES, facecolor='#d9eaf7', edgecolor='gray')

# --- 4. Add Latitude and Longitude Gridlines ---
gl = ax.gridlines(crs=transform, draw_labels=True,
                  linewidth=1, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 12, 'color': 'black'}
gl.ylabel_style = {'size': 12, 'color': 'black'}

# --- 5. Plot the Stations ---
ax.scatter(df_stations['lon'], df_stations['lat'], 
           s=120,
           color='#c00000',
           edgecolor='black',
           linewidth=1,
           label='GNSS Stations',
           transform=transform,
           zorder=10)

# --- 6. Add Station Labels ---
for station_name, coords in df_stations.iterrows():
    ax.text(coords['lon'] + 3, coords['lat'], station_name,
            transform=transform,
            ha='left',
            va='center',
            fontsize=11, 
            fontweight='bold',
            color='black',
            bbox=dict(boxstyle="round,pad=0.2", facecolor='white', alpha=0.7, edgecolor='none'))

# --- 7. Final Touches ---
ax.legend(loc='lower left', fontsize=14, frameon=True, fancybox=True, shadow=True)

# --- 8. Save the Figure ---
plt.savefig('/home/akash/Desktop/VTEC-May-2024-Final-Year-Project-2025-/PLOTS/station_map_final.pdf', bbox_inches='tight')

plt.show()
```
