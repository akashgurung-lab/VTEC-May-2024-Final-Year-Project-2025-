```python
# ------- LOADING PACKAGES --------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.patches import Rectangle
from datetime import datetime
import seaborn as sns
```

```python
# -------------Defining path---------------
path_omni_hourly = '/home/akash/Desktop/VTEC-May-2024-Final-Year-Project-2025-/Data/OMNI_PARAMETERS/solar_parameters.txt'
path_symh_5min = '/home/akash/Desktop/VTEC-May-2024-Final-Year-Project-2025-/Data/OMNI_PARAMETERS/SYM-H.txt'
```

```python
# Set up journal-quality plotting parameters with cross-platform font compatibility
plt.rcParams.update({
    'font.size': 11,
    'font.family': 'serif',
    'font.serif': ['DejaVu Serif', 'Liberation Serif', 'Times', 'serif'],
    'axes.linewidth': 1.2,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.dpi': 150,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight',
    'text.usetex': False,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'grid.linewidth': 0.5
})
# Define color palette for scientific publication
colors = {
    'bz': '#d62728',          # Red for Bz
    'speed': '#1f77b4',       # Blue for solar wind speed
    'pressure': '#ff7f0e',    # Orange for pressure
    'density': '#2ca02c',     # Green for density
    'temp': '#9467bd',        # Purple for temperature
    'symh': '#17becf',        # Teal for SYM-H 5-min
    'symh_smooth': '#8B0000', # Dark red for SYM-H 1-hr average
    'storm': '#ffcccc',       # Light red for storm period
    'reference': '#666666'    # Gray for reference lines
}
# --- Define Fill Values for Missing Data ---
omni_fill_values = [9999999., 999.9, 9999., 99.99]
symh_fill_values = [99999]

try:
    # --- 1. Load OMNI Data (HOURLY) ---
    omni_hourly_df = pd.read_csv(
        path_omni_hourly,
        skiprows=9,
        sep='\s+',
        names=['Year', 'DOY', 'Hour', 'Bz_GSM', 'Temp', 'Proton_Density', 'Flow_Speed', 'Flow_Pressure'],
        na_values=omni_fill_values
    )
    
    # Create datetime index
    base_date_hourly = pd.to_datetime(omni_hourly_df['Year'], format='%Y')
    omni_hourly_df['datetime'] = (base_date_hourly + 
                                 pd.to_timedelta(omni_hourly_df['DOY'] - 1, unit='D') + 
                                 pd.to_timedelta(omni_hourly_df['Hour'], unit='H'))
    omni_hourly_df = omni_hourly_df.set_index('datetime')

    # --- 2. Load SYM-H Data (5-MINUTE) ---
    symh_5min_df = pd.read_csv(
        path_symh_5min,
        skiprows=5,
        sep='\s+',
        names=['Year', 'DOY', 'Hour', 'Minute', 'SYM-H'],
        na_values=symh_fill_values
    )
    
    # Create datetime index
    base_date_5min = pd.to_datetime(symh_5min_df['Year'], format='%Y')
    symh_5min_df['datetime'] = (base_date_5min + 
                               pd.to_timedelta(symh_5min_df['DOY'] - 1, unit='D') + 
                               pd.to_timedelta(symh_5min_df['Hour'], unit='H') + 
                               pd.to_timedelta(symh_5min_df['Minute'], unit='min'))
    symh_5min_df = symh_5min_df.set_index('datetime')

    print("â†’ Data loaded successfully!")
    
    # --- 3. Advanced NaN Handling for OMNI data ---
    def hybrid_nan_handling(series, param_name):
        """Hybrid NaN handling approach optimized for space weather data"""
        original_series = series.copy()
        
        if param_name.lower() in ['bz_gsm', 'bz']:
            series = series.interpolate(method='linear', limit=2)
            series = series.fillna(0, limit=4)
            
        elif param_name.lower() in ['flow_speed', 'vsw', 'speed']:
            series = series.interpolate(method='linear', limit=4)
            series = series.ffill(limit=3)
            series = series.fillna(400)
            
        elif param_name.lower() in ['proton_density', 'density', 'np']:
            series = series.interpolate(method='linear', limit=3)
            rolling_median = series.rolling(window=24, center=True, min_periods=6).median()
            series = series.fillna(rolling_median)
            series = series.fillna(5.0)
            
        elif param_name.lower() in ['temp', 'temperature']:
            series = series.interpolate(method='linear', limit=4)
            rolling_mean = series.rolling(window=48, center=True, min_periods=12).mean()
            series = series.fillna(rolling_mean)
            series = series.fillna(100000)
            
        elif param_name.lower() in ['flow_pressure', 'pressure', 'psw']:
            series = series.interpolate(method='linear', limit=3)
            series = series.ffill(limit=4)
            series = series.fillna(2.0)
        
        return series
    
    # Apply NaN handling to OMNI data
    processed_omni = {}
    for column in omni_hourly_df.columns:
        if column not in ['Year', 'DOY', 'Hour']:
            processed_omni[column] = hybrid_nan_handling(omni_hourly_df[column], column)
    
    final_omni_df = pd.DataFrame(processed_omni, index=omni_hourly_df.index)
    
    # Handle SYM-H NaN values minimally
    symh_5min_df['SYM-H'] = symh_5min_df['SYM-H'].interpolate(method='linear', limit=2)
    
    # --- 4. Create Simplified Figure with 6 panels ---
    fig, axes = plt.subplots(6, 1, figsize=(14, 16), sharex=True)
    fig.subplots_adjust(hspace=0.15)

    # Define storm period
    storm_start = pd.Timestamp('2024-05-10 17:00')
    storm_end = pd.Timestamp('2024-05-12 12:00')

    # --- Panel (a): Bz GSM with adjusted y-limits ---
    axes[0].plot(final_omni_df.index, final_omni_df['Bz_GSM'], 
                 color=colors['bz'], linewidth=1.5, alpha=0.9, label=r'$B_z$ GSM')
    axes[0].axhline(0, color=colors['reference'], linestyle='--', linewidth=1, alpha=0.7)
    # The 'Storm threshold' line below has been removed.
    # axes[0].axhline(-10, color='red', linestyle=':', linewidth=1, alpha=0.7, label='Storm threshold') 
    axes[0].set_ylabel(r'$B_z$ GSM (nT)', fontweight='bold')
    axes[0].set_ylim(-40, 20)
    axes[0].axvspan(storm_start, storm_end, color=colors['storm'], alpha=0.3, zorder=0)
    axes[0].text(0.02, 0.92, '(a)', transform=axes[0].transAxes, fontsize=14, 
                 fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    axes[0].legend(loc='upper right', framealpha=0.9, fontsize=9)

    # --- Panel (b): Solar Wind Speed ---
    axes[1].plot(final_omni_df.index, final_omni_df['Flow_Speed'], 
                 color=colors['speed'], linewidth=1.5, alpha=0.9, label=r'$V_{sw}$')
    axes[1].set_ylabel(r'$V_{sw}$ (km s$^{-1}$)', fontweight='bold')
    axes[1].set_ylim(200, 1200)
    axes[1].axvspan(storm_start, storm_end, color=colors['storm'], alpha=0.3, zorder=0)
    axes[1].text(0.02, 0.92, '(b)', transform=axes[1].transAxes, fontsize=14, 
                 fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    axes[1].legend(loc='upper right', framealpha=0.9, fontsize=9)

    # --- Panel (c): Solar Wind Pressure ---
    axes[2].plot(final_omni_df.index, final_omni_df['Flow_Pressure'], 
                 color=colors['pressure'], linewidth=1.5, alpha=0.9, label=r'$P_{sw}$')
    axes[2].set_ylabel(r'$P_{sw}$ (nPa)', fontweight='bold')
    axes[2].set_ylim(0, 60)
    axes[2].axvspan(storm_start, storm_end, color=colors['storm'], alpha=0.3, zorder=0)
    axes[2].text(0.02, 0.92, '(c)', transform=axes[2].transAxes, fontsize=14, 
                 fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    axes[2].legend(loc='upper right', framealpha=0.9, fontsize=9)

    # --- Panel (d): Proton Density ---
    axes[3].plot(final_omni_df.index, final_omni_df['Proton_Density'], 
                 color=colors['density'], linewidth=1.5, alpha=0.9, label=r'$N_p$')
    axes[3].set_ylabel(r'$N_p$ (cm$^{-3}$)', fontweight='bold')
    axes[3].set_ylim(0, 100)
    axes[3].axvspan(storm_start, storm_end, color=colors['storm'], alpha=0.3, zorder=0)
    axes[3].text(0.02, 0.92, '(d)', transform=axes[3].transAxes, fontsize=14, 
                 fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    axes[3].legend(loc='upper right', framealpha=0.9, fontsize=9)

    # --- Panel (e): Temperature ---
    axes[4].plot(final_omni_df.index, final_omni_df['Temp'], 
                 color=colors['temp'], linewidth=1.5, alpha=0.9, label='Temperature')
    axes[4].set_ylabel('Temperature (K)', fontweight='bold')
    axes[4].set_yscale('log')
    axes[4].axvspan(storm_start, storm_end, color=colors['storm'], alpha=0.3, zorder=0)
    axes[4].text(0.02, 0.92, '(e)', transform=axes[4].transAxes, fontsize=14, 
                 fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))
    axes[4].legend(loc='upper right', framealpha=0.9, fontsize=9)

    # --- Panel (f): SYM-H ---
    axes[5].plot(symh_5min_df.index, symh_5min_df['SYM-H'], 
                 color=colors['symh'], linewidth=1.5, alpha=0.9, 
                 label='SYM-H', zorder=1)
    
    axes[5].axhline(0, color=colors['reference'], linestyle='--', linewidth=1, alpha=0.7)
    # The storm classification lines below have been removed.
    # axes[5].axhline(-50, color='orange', linestyle=':', linewidth=1, alpha=0.7, label='Moderate storm')
    # axes[5].axhline(-100, color='red', linestyle=':', linewidth=1, alpha=0.7, label='Intense storm')
    # axes[5].axhline(-200, color='darkred', linestyle=':', linewidth=1, alpha=0.7, label='Super storm')
    
    axes[5].set_ylabel('SYM-H (nT)', fontweight='bold')
    axes[5].axvspan(storm_start, storm_end, color=colors['storm'], alpha=0.3, zorder=0)
    axes[5].legend(loc='lower right', framealpha=0.9, fontsize=9)
    axes[5].text(0.02, 0.92, '(f)', transform=axes[5].transAxes, fontsize=14, 
                 fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

    # --- Format all axes ---
    axes[5].set_xlabel('Date (May 2024, UT)', fontweight='bold')
    axes[5].xaxis.set_major_locator(mdates.DayLocator(interval=1))
    axes[5].xaxis.set_minor_locator(mdates.HourLocator(interval=6))
    axes[5].xaxis.set_major_formatter(mdates.DateFormatter('%d'))
    
    # Set consistent formatting for all panels
    for ax in axes:
        ax.set_xlim([final_omni_df.index.min(), final_omni_df.index.max()])
        ax.grid(True, alpha=0.3, linewidth=0.5)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(1.2)
        ax.spines['bottom'].set_linewidth(1.2)

    
    # Add annotation for storm center
    storm_center = storm_start + (storm_end - storm_start) / 2
    axes[0].annotate('Intense\nStorm Period', 
                     xy=(storm_center, 15), 
                     ha='center', va='center', 
                     fontsize=10, fontweight='bold',
                     bbox=dict(boxstyle="round,pad=0.4", facecolor='white', 
                              alpha=0.9, edgecolor='red', linewidth=1))

    # Save high-quality figures
    plt.savefig('simplified_space_weather_analysis.png', dpi=300, bbox_inches='tight', 
                facecolor='white', edgecolor='none')
    plt.savefig('simplified_space_weather_analysis.pdf', bbox_inches='tight', 
                facecolor='white', edgecolor='none')
    
    plt.tight_layout()
    plt.show()
    
    print("â†’ Simplified multi-parameter analysis completed!")
    print("â†’ Six panels created:")
    print("  â€¢ (a) Bz GSM with extended range (-40 to +20 nT)")
    print("  â€¢ (b) Solar wind speed")
    print("  â€¢ (c) Solar wind pressure") 
    print("  â€¢ (d) Proton density")
    print("  â€¢ (e) Temperature (log scale)")
    print("  â€¢ (f) SYM-H (5-minute resolution only)")
    print("â†’ Files saved: simplified_space_weather_analysis.png/.pdf")

except FileNotFoundError:
    print("â†’ ERROR: Data files not found. Please check your file paths.")
except Exception as e:
    print(f"â†’ ERROR: {e}")
```

```python
# --- ANALYSIS BLOCK FOR QUIET DAYS---

print("\n" + "="*50)
print("ANALYSIS OF THE QUIET PERIOD")
print("="*50)

# 1. Define the exact time range for the quiet period
# We'll stop just before the storm hits around 17:00 UT on May 10
start_date = pd.Timestamp('2024-05-07 00:00')
end_date = pd.Timestamp('2024-05-10 16:00')

# 2. Create sliced DataFrames for just this period
quiet_omni_df = final_omni_df.loc[start_date:end_date]
quiet_symh_df = symh_5min_df.loc[start_date:end_date]

# --- 3. Analyze each parameter ---

# (a) IMF Bz Analysis
print("\n--- Panel (a): IMF Bz ---")
bz_min = quiet_omni_df['Bz_GSM'].min()
bz_max = quiet_omni_df['Bz_GSM'].max()
print(f"Bz fluctuated between {bz_min:.1f} nT and {bz_max:.1f} nT.")

# (b) Solar Wind Speed Analysis
print("\n--- Panel (b): Solar Wind Speed (Vsw) ---")
speed_mean = quiet_omni_df['Flow_Speed'].mean()
speed_min = quiet_omni_df['Flow_Speed'].min()
speed_max = quiet_omni_df['Flow_Speed'].max()
print(f"Vsw ranged from {speed_min:.1f} to {speed_max:.1f} km/s, with a mean of {speed_mean:.1f} km/s.")

# (c) Solar Wind Pressure Analysis
print("\n--- Panel (c): Solar Wind Pressure (Psw) ---")
pressure_mean = quiet_omni_df['Flow_Pressure'].mean()
pressure_95th_percentile = quiet_omni_df['Flow_Pressure'].quantile(0.95)
print(f"Psw had a mean of {pressure_mean:.2f} nPa, with 95% of values below {pressure_95th_percentile:.2f} nPa.")

# (d) Proton Density Analysis --- ADDED ---
print("\n--- Panel (d): Proton Density (Np) ---")
density_mean = quiet_omni_df['Proton_Density'].mean()
density_min = quiet_omni_df['Proton_Density'].min()
density_max = quiet_omni_df['Proton_Density'].max()
print(f"Np ranged from {density_min:.1f} to {density_max:.1f} N/cm^3, with a mean of {density_mean:.1f} N/cm^3.")

# (e) Temperature Analysis --- ADDED ---
print("\n--- Panel (e): Temperature ---")
temp_mean = quiet_omni_df['Temp'].mean()
# Use scientific notation for large numbers to make it readable
print(f"Temperature had a mean value of {temp_mean:.2e} K.")

# (f) SYM-H Analysis
print("\n--- Panel (f): SYM-H ---")
symh_min = quiet_symh_df['SYM-H'].min()
symh_max = quiet_symh_df['SYM-H'].max()
symh_mean = quiet_symh_df['SYM-H'].mean()
print(f"SYM-H ranged from {symh_min:.1f} nT to {symh_max:.1f} nT, with a mean of {symh_mean:.1f} nT.")

print("\n" + "="*50 + "\n")

# Your existing plotting code would continue after this block...
```

```python
# --- ANALYSIS BLOCK FOR THE STORM COMMENCEMENT ---

print("\n" + "="*50)
print("ANALYSIS OF THE SHOCK AND COMMENCEMENT PERIOD")
print("="*50)

# 1. Define the exact time range for this phase
# We'll look from just before the shock to the end of the day on May 10
start_date_shock = pd.Timestamp('2024-05-10 16:00')
end_date_shock = pd.Timestamp('2024-05-10 23:59')

# 2. Create sliced DataFrames for this period
shock_omni_df = final_omni_df.loc[start_date_shock:end_date_shock]
shock_symh_df = symh_5min_df.loc[start_date_shock:end_date_shock]

# --- 3. Analyze each parameter during the commencement ---

# (a) IMF Bz Analysis
print("\n--- Panel (a): IMF Bz ---")
pre_shock_bz = final_omni_df.loc['2024-05-10 16:00']['Bz_GSM']
min_bz_after_shock = shock_omni_df['Bz_GSM'].min()
print(f"Pre-shock Bz was {pre_shock_bz:.1f} nT.")
print(f"After the shock, Bz turned sharply south, reaching a minimum of {min_bz_after_shock:.1f} nT on May 10.")

# (b) Solar Wind Speed Analysis
print("\n--- Panel (b): Solar Wind Speed (Vsw) ---")
pre_shock_speed = final_omni_df.loc['2024-05-10 16:00']['Flow_Speed']
max_speed = shock_omni_df['Flow_Speed'].max()
print(f"Vsw jumped from a pre-shock value of {pre_shock_speed:.1f} km/s to a peak of {max_speed:.1f} km/s.")

# (c) Solar Wind Pressure Analysis
print("\n--- Panel (c): Solar Wind Pressure (Psw) ---")
pre_shock_pressure = final_omni_df.loc['2024-05-10 16:00']['Flow_Pressure']
max_pressure = shock_omni_df['Flow_Pressure'].max()
print(f"Psw jumped from a pre-shock value of {pre_shock_pressure:.2f} nPa to a peak of {max_pressure:.2f} nPa.")

# (d) Proton Density Analysis
print("\n--- Panel (d): Proton Density (Np) ---")
pre_shock_density = final_omni_df.loc['2024-05-10 16:00']['Proton_Density']
max_density = shock_omni_df['Proton_Density'].max()
print(f"Np jumped from a pre-shock value of {pre_shock_density:.1f} N/cm^3 to a peak of {max_density:.1f} N/cm^3.")

# (e) Temperature Analysis
print("\n--- Panel (e): Temperature ---")
pre_shock_temp = final_omni_df.loc['2024-05-10 16:00']['Temp']
max_temp = shock_omni_df['Temp'].max()
print(f"Temperature jumped from a pre-shock value of {pre_shock_temp:.2e} K to a peak of {max_temp:.2e} K.")

# (f) SYM-H Analysis
print("\n--- Panel (f): SYM-H ---")
# Find the value right before the jump
pre_ssc_symh = symh_5min_df.loc['2024-05-10 17:05']['SYM-H']
# Find the peak of the positive jump (the SSC)
ssc_peak_symh = shock_symh_df['SYM-H'].max()
# Find the minimum value reached as the main phase began on this day
main_phase_start_min_symh = shock_symh_df['SYM-H'].min()
print(f"SYM-H shows a Storm Sudden Commencement (SSC), jumping from {pre_ssc_symh:.1f} nT to a peak of {ssc_peak_symh:.1f} nT.")
print(f"Following the southward turning, SYM-H began to plunge, reaching {main_phase_start_min_symh:.1f} nT by the end of May 10.")

print("\n" + "="*50 + "\n")
```

```python
# --- ANALYSIS BLOCK FOR THE MAIN STORM PHASE ---

print("\n" + "="*50)
print("ANALYSIS OF THE MAIN STORM PHASE")
print("="*50)

# 1. Define the exact time range for the Main Phase (all of May 11)
start_date_main = pd.Timestamp('2024-05-11 00:00')
end_date_main = pd.Timestamp('2024-05-11 23:59')

# 2. Create sliced DataFrames for this period
main_omni_df = final_omni_df.loc[start_date_main:end_date_main]
main_symh_df = symh_5min_df.loc[start_date_main:end_date_main]

# --- 3. Analyze each parameter during the Main Phase ---

# (a) IMF Bz Analysis
print("\n--- Panel (a): IMF Bz ---")
min_bz_main = main_omni_df['Bz_GSM'].min()
mean_bz_main = main_omni_df['Bz_GSM'].mean()
print(f"The IMF Bz remained strongly southward, reaching a minimum of {min_bz_main:.1f} nT.")
print(f"The average Bz during the main phase was {mean_bz_main:.1f} nT, indicating sustained forcing.")

# (b) Solar Wind Speed Analysis
print("\n--- Panel (b): Solar Wind Speed (Vsw) ---")
mean_speed_main = main_omni_df['Flow_Speed'].mean()
max_speed_main = main_omni_df['Flow_Speed'].max()
print(f"Solar wind speed remained elevated, averaging {mean_speed_main:.1f} km/s and peaking at {max_speed_main:.1f} km/s.")

# (c) Solar Wind Pressure Analysis
print("\n--- Panel (c): Solar Wind Pressure (Psw) ---")
mean_pressure_main = main_omni_df['Flow_Pressure'].mean()
max_pressure_main = main_omni_df['Flow_Pressure'].max()
print(f"Dynamic pressure was highly variable, with a mean of {mean_pressure_main:.2f} nPa and a peak of {max_pressure_main:.2f} nPa.")

# (d) Proton Density Analysis
print("\n--- Panel (d): Proton Density (Np) ---")
mean_density_main = main_omni_df['Proton_Density'].mean()
max_density_main = main_omni_df['Proton_Density'].max()
print(f"Proton density had a mean of {mean_density_main:.1f} N/cm^3 and peaked at {max_density_main:.1f} N/cm^3.")

# (e) Temperature Analysis
print("\n--- Panel (e): Temperature ---")
mean_temp_main = main_omni_df['Temp'].mean()
max_temp_main = main_omni_df['Temp'].max()
print(f"Plasma temperature was extremely high, averaging {mean_temp_main:.2e} K and peaking at {max_temp_main:.2e} K.")

# (f) SYM-H Analysis
print("\n--- Panel (f): SYM-H ---")
min_symh_main = main_symh_df['SYM-H'].min()
time_of_min_symh = main_symh_df['SYM-H'].idxmin()
print(f"The storm reached its peak intensity with a minimum SYM-H of {min_symh_main:.1f} nT.")
print(f"This minimum occurred at {time_of_min_symh.strftime('%Y-%m-%d %H:%M')} UT.")

print("\n" + "="*50 + "\n")
```

```python
# --- ANALYSIS BLOCK FOR THE RECOVERY PHASE ---

print("\n" + "="*50)
print("ANALYSIS OF THE RECOVERY PHASE")
print("="*50)

# 1. Define the time range for the Recovery Phase (May 12 to end of data)
start_date_recovery = pd.Timestamp('2024-05-12 00:00')
end_date_recovery = pd.Timestamp('2024-05-15 23:59')

# 2. Create sliced DataFrames for this period
recovery_omni_df = final_omni_df.loc[start_date_recovery:end_date_recovery]
recovery_symh_df = symh_5min_df.loc[start_date_recovery:end_date_recovery]

# --- 3. Analyze each parameter during the Recovery Phase ---

# (a) IMF Bz Analysis
print("\n--- Panel (a): IMF Bz ---")
mean_bz_recovery = recovery_omni_df['Bz_GSM'].mean()
print(f"The IMF Bz forcing subsided, with the average Bz returning to a near-neutral value of {mean_bz_recovery:.1f} nT.")

# (b) Solar Wind Speed Analysis
print("\n--- Panel (b): Solar Wind Speed (Vsw) ---")
mean_speed_recovery = recovery_omni_df['Flow_Speed'].mean()
final_speed_recovery = recovery_omni_df['Flow_Speed'].iloc[-1] # .iloc[-1] gets the last value
print(f"Solar wind speed gradually decreased, averaging {mean_speed_recovery:.1f} km/s and ending the period at {final_speed_recovery:.1f} km/s.")

# (c) Solar Wind Pressure Analysis
print("\n--- Panel (c): Solar Wind Pressure (Psw) ---")
mean_pressure_recovery = recovery_omni_df['Flow_Pressure'].mean()
print(f"Dynamic pressure decreased significantly, with a mean of {mean_pressure_recovery:.2f} nPa during the recovery.")

# (d) Proton Density Analysis
print("\n--- Panel (d): Proton Density (Np) ---")
mean_density_recovery = recovery_omni_df['Proton_Density'].mean()
print(f"Proton density returned to lower levels, with a mean of {mean_density_recovery:.1f} N/cm^3.")

# (e) Temperature Analysis
print("\n--- Panel (e): Temperature ---")
mean_temp_recovery = recovery_omni_df['Temp'].mean()
print(f"Plasma temperature cooled considerably, averaging {mean_temp_recovery:.2e} K.")

# (f) SYM-H Analysis
print("\n--- Panel (f): SYM-H ---")
symh_start_recovery = recovery_symh_df['SYM-H'].iloc[0] # .iloc[0] gets the first value
symh_end_recovery = recovery_symh_df['SYM-H'].iloc[-1]
print(f"The SYM-H index began the recovery phase at {symh_start_recovery:.1f} nT on May 12.")
print(f"By the end of May 15, it had recovered to {symh_end_recovery:.1f} nT, but had not yet reached its quiet-time baseline.")

print("\n" + "="*50 + "\n")

# Your existing plotting code would continue after this block...
```
