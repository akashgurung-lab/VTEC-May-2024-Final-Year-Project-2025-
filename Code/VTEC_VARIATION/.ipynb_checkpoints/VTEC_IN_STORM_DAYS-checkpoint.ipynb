```python
# importing the packages 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates 
```

```python
# Cell 2: Load and Merge
storm_data_list = []

print("Loading and labeling all storm day files...")
for station_name, file_list in files_to_process_storm.items():
    for path, day_label in file_list:
        try:
            df = pd.read_csv(path, skiprows=4, sep='\s+')
            df = df[df['Time'] >= 0].dropna()
            
            df['Station'] = station_name
            df['Day'] = day_label
            
            storm_data_list.append(df)
        except FileNotFoundError:
            print(f"--> WARNING: File not found and will be skipped: {path}")
        except Exception as e:
            print(f"--> ERROR: Could not process {path}. Reason: {e}")

if not storm_data_list:
    print("No storm data was loaded. Cannot continue.")
else:
    storm_master_df = pd.concat(storm_data_list, ignore_index=True)
    print("\nAll storm day files successfully merged into 'storm_master_df'.")
    print(f"Total rows loaded: {len(storm_master_df)}")
```

```python
# Cell 3: Final Preparation
if 'storm_master_df' in locals():
    storm_master_df['datetime'] = pd.to_datetime(storm_master_df['MJdatet'], unit='D', origin='1858-11-17')
    storm_master_df = storm_master_df.sort_values(by='datetime')
    
    print("Final storm data preparation complete. 'storm_master_df' is sorted and ready for plotting.")
    storm_master_df.info()
else:
    print("'storm_master_df' not found. Please run the previous cell first.")
```

```python
# Cell 4: Plotting 3 Stations per Subplot for Comparison
if 'storm_master_df' in locals():
    stations_to_plot = storm_master_df['Station'].unique()
    
    # Create 3x1 subplot grid
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(20, 18), sharex=True)
    
    # Define colors for the 3 stations in each subplot
    colors = ['blue', 'orange', 'purple']
    
    print("Generating comparative storm day plots...")
    
    # Group stations into sets of 3
    for i, ax in enumerate(axes):
        # Get 3 stations for this subplot
        start_idx = i * 3
        end_idx = min(start_idx + 3, len(stations_to_plot))
        current_stations = stations_to_plot[start_idx:end_idx]
        
        # Plot each station in the current subplot
        for j, station_name in enumerate(current_stations):
            station_df = storm_master_df[storm_master_df['Station'] == station_name]
            resampled_data = station_df.set_index('datetime')['Vtec'].resample('15T').mean()
            
            # Plot with different colors for each station
            ax.plot(resampled_data.index, resampled_data.values, 
                   label=f'{station_name}', color=colors[j], linewidth=2.5)
        
        # Add day boundaries (using data from first station for reference)
        if len(current_stations) > 0:
            first_station_df = storm_master_df[storm_master_df['Station'] == current_stations[0]]
            first_resampled = first_station_df.set_index('datetime')['Vtec'].resample('15T').mean()
            
            if not first_resampled.empty:
                day_boundaries = pd.to_datetime(first_resampled.index.date).unique()[1:]
                for boundary in day_boundaries:
                    ax.axvline(x=boundary, color='black', linestyle='--', linewidth=1.5, alpha=0.6)
                    
                unique_days = pd.to_datetime(first_resampled.index.date).unique()
                y_max = ax.get_ylim()[1]
                y_pos = y_max * 0.9
                
                for day in unique_days:
                    pos = pd.to_datetime(day) + pd.Timedelta(hours=12)
                    text = day.strftime('%b %d')
                    ax.text(pos, y_pos, text, ha='center', fontsize=16, 
                           fontweight='bold', color='gray')
        
        # Format x-axis
        ax.xaxis.set_major_locator(mdates.HourLocator(interval=6))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        
        # Set labels and formatting
        ax.set_title(f"Comparison: {' vs '.join(current_stations)}", fontsize=20, weight='bold')
        ax.set_ylabel("Averaged VTEC (TECU)", fontsize=16)
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.legend(fontsize=14, loc='upper right')
        ax.tick_params(axis='both', which='major', labelsize=12)
    
    # Add main labels
    fig.text(0.5, 0.02, 'Time (UT)', ha='center', va='center', fontsize=18, weight='bold')
    fig.suptitle("VTEC Comparison During Geomagnetic Storm (May 10-12, 2024)", 
                 fontsize=24, weight='bold')
    
    plt.tight_layout(rect=[0, 0.05, 1, 0.96])
    plt.savefig('/home/akash/Desktop/VTEC-May-2024-Final-Year-Project-2025-/PLOTS/vtec_plot_for_storm_day.pdf', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Print station groupings for reference
    print("\nStation groupings:")
    for i in range(3):
        start_idx = i * 3
        end_idx = min(start_idx + 3, len(stations_to_plot))
        current_stations = stations_to_plot[start_idx:end_idx]
        print(f"Subplot {i+1}: {', '.join(current_stations)}")
        
else:
    print("'storm_master_df' not found. Please run the previous cells first.")
```
